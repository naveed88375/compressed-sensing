```
rate, y = siw.read('yow1.wav')
n = len(y)
t = np.linspace(0, n/rate, n)
yt = spfft.dct(y, norm='ortho')
plt.subplot(121)
plt.rcParams['figure.figsize']=(40,10)
#plt.xlim(0,0.008)
plt.plot(t,y)
plt.subplot(122)
plt.plot(yt)
#plt.xlim(0,600)
```

```
# extract small sample of signal
m = int(n*(3/10)) # 20% sample
a1 = -5*np.random.rand(n)
a2 =  5*np.random.rand(n)
a = toeplitz(a1,a2)
ri = np.random.choice(n, m, replace=False)
ri.sort()
ri = a[ri]
y2 = np.dot(ri,y)
```

```
# create idct matrix operator
A = spfft.idct(np.identity(n), norm='ortho', axis=0)
A = np.dot(ri,A)

# do L1 optimization
vx = cvx.Variable(n)
objective = cvx.Minimize(cvx.norm(vx, 1))
constraints = [A*vx == y2]
prob = cvx.Problem(objective, constraints)
result = prob.solve(verbose=True)
```

```
# reconstruct signal
x = np.array(vx.value)
x = np.squeeze(x)
sig = spfft.idct(x, norm='ortho', axis=0)
plt.plot(t,sig)
plt.rcParams['figure.figsize']=(10,10)
#plt.xlim(0.1,0.3)
```
